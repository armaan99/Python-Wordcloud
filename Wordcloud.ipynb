# WORDCLOUD

Word clouds or tag clouds are graphical representations of word frequency that give greater prominence to words that appear more frequently in a source text. The larger the word in the visual the more common the word was in the document(s). This type of visualization can assist evaluators with exploratory textual analysis by identifying words that frequently appear in a set of interviews, documents, or other text. It can also be used for communicating the most salient points or themes in the reporting stage.

```python
# importing wordcloud package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud imported!')
```

<b>Let's try to analyze a short novel written by Lewis Carroll titled Alice's Adventures in Wonderland.</b>

```python
# Fetching txt file (saved in same location as "novel.txt")

novel =  open("novel.txt","r").read()
print ('File read!')
```

```python
# let's use the stopwords that we imported from wordcloud.
# We use the function set to remove any redundant stopwords.

stopwords = set(STOPWORDS)

print(stopwords)
```

```python
# Some words are not in stopwords like "said"

print('said' in stopwords)
```

```python
# You may include more words in this set manually according to your need
# Like we have added "said" in this set

stopwords.add('said')
print(stopwords)
```

```python
# Now you can verify that the words we added manually gets appended in the set of stopwords

print('said' in stopwords)
```

```python
# Removing Punctuations from the text

punctuations = '''!()-[]{'"};:\\<,>./?@#$&^%*_~'''

for letter in novel:
    if letter in punctuations:
        novel = novel.replace(letter,"")
```

```python
# Creating dictionary or words with frequence, excluding stopwords

words = dict()
for word in novel.split():
    if word.lower() not in stopwords:
        words[word] = words.get(word,0) + 1
```

```python
# Instantiate a WordCloud object

wc = WordCloud(
    background_color='black',
    max_words=4000,
    width = 400,
    height = 200
)
```

```python
# generate the wordcloud

wc.generate_from_frequencies(words)
```

```python
# importing matplotlib.pyplot to display wordcloud
import matplotlib.pyplot as plt

print("matplotlib.pyplot imported!")
```

```python
# display the wordcloud

plt.imshow(wc, interpolation='bilinear')
'''
interpolation can have these set of values :
{'nearest', 'bilinear', 'bicubic', 'spline16', 'spline36', 'hanning', 'hamming', 'hermite',
'kaiser','quadric','catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos', 'none'}
'''
plt.axis('off')
plt.show()
```

<b>Creating wordcloud on Masked Image</b>

```python
# importing Image to fetch image
from PIL import Image

print("Image imported!")
```

```python
# importing numpy for image processing
import numpy as np

print("numpy imported!")
```

```python
mask_img = np.array(Image.open('mask_image.jpg'))

fig = plt.figure()
fig.set_figwidth(4) # set width
fig.set_figheight(4) # set height

plt.imshow(mask_img, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Instantiate a WordCloud object

wc_mask = WordCloud(
    background_color='black',
    max_words=4000,
    mask = mask_img,
)
```

```python
# generate the wordcloud

wc_mask.generate_from_frequencies(words)
```

```python
# display the wordcloud

plt.imshow(wc_mask, interpolation='bilinear')
plt.axis('off')
plt.show()
```
